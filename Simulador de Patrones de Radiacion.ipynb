# MAPAS POLARES

```python
import tkinter as tk
from tkinter import ttk
import numpy as np
import matplotlib.pyplot as plt

def plot_graph():
    f = float(entry_frecuencia.get())
    c = float(entry_velocidad.get())
    d = float(entry_d.get())
    A = float(entry_A.get())
    R = float(entry_R.get())
    omega = 2 * np.pi * f
    k = omega / c
    Q = np.linspace(0, 2 * np.pi, 360)  # Ángulo en radianes

    
    AMP_monopolo = (A / R) * np.exp(-1j * k * R)
    AMP_dipolo = (A / R) * (2j * np.sin(k * (d / 2) * np.sin(Q)))
    AMP_tripoloneg = (A / R) * (-1 + 2 * np.cos(k * d * np.sin(Q)))
    AMP_tripolopos = (A / R) * (1 - 2 * np.cos(k * d * np.sin(Q)))
    AMP_cuadripolo = (A / R) * (
        (7/8) * np.cos((3/2) * k * d * np.sin(Q)) -
        (1/4) * np.cos(k * (d/2) * np.sin(Q)) +
        2j * np.sin((3/2) * k * d * np.sin(Q)) -
        (3/4) * 1j * np.sin(k * (d/2) * np.sin(Q))
    )

    
    plt.figure(figsize=(12, 10))
    
    tipo_grafica = combo_tipo_grafica.get()
    
    if tipo_grafica == "Monopolo":
        plt.subplot(1, 1, 1, polar=True)
        plt.title('Monopolo')
        plt.polar(Q, np.abs(AMP_monopolo) * np.ones_like(Q))
    
    elif tipo_grafica == "Dipolo":
        plt.subplot(1, 1, 1, polar=True)
        plt.title('Dipolo Positivo')
        plt.polar(Q, np.abs(AMP_dipolo))
    
    elif tipo_grafica == "Tripolo positivo":
        plt.subplot(1, 1, 1, polar=True)
        plt.title('Tripolo Positivo')
        plt.polar(Q, np.abs(AMP_tripolopos))
    
    elif tipo_grafica == "Tripolo negativo":
        plt.subplot(1, 1, 1, polar=True)
        plt.title('Tripolo Negativo')
        plt.polar(Q, np.abs(AMP_tripoloneg))
    
    elif tipo_grafica == "Cuadripolo":
        plt.subplot(1, 1, 1, polar=True)
        plt.title('Cuadripolo')
        plt.polar(Q, np.abs(AMP_cuadripolo))

    plt.tight_layout()
    plt.show()


root = tk.Tk()
root.title("Interfaz Gráfica para Gráficas de Antenas")


ttk.Label(root, text="Frecuencia (Hz):").grid(column=0, row=0)
entry_frecuencia = ttk.Entry(root)
entry_frecuencia.grid(column=1, row=0)

ttk.Label(root, text="Velocidad de Propagación (m/s):").grid(column=0, row=1)
entry_velocidad = ttk.Entry(root)
entry_velocidad.grid(column=1, row=1)

ttk.Label(root, text="Separación del dipolo/tripolo/cuadripolo (m):").grid(column=0, row=2)
entry_d = ttk.Entry(root)
entry_d.grid(column=1, row=2)

ttk.Label(root, text="Amplitud:").grid(column=0, row=3)
entry_A = ttk.Entry(root)
entry_A.grid(column=1, row=3)

ttk.Label(root, text="Distancia al punto del oyente(m):").grid(column=0, row=4)
entry_R = ttk.Entry(root)
entry_R.grid(column=1, row=4)


ttk.Label(root, text="Tipo de Gráfica:").grid(column=0, row=5)
combo_tipo_grafica = ttk.Combobox(root, values=["Monopolo", "Dipolo", "Tripolo positivo", "Tripolo negativo", "Cuadripolo"])
combo_tipo_grafica.grid(column=1, row=5)
combo_tipo_grafica['state'] = 'readonly'  


style = ttk.Style()
style.configure("TButton", foreground="black")


btn_plot = ttk.Button(root, text="Graficar", command=plot_graph, style="TButton")
btn_plot.grid(column=0, row=6, columnspan=2)


root.mainloop()
```

# Parámetros del medio y la fuente

f = 100    # Frecuencia (Hz)

c = 343    # Velocidad del sonido (m/s)

d = 0.5    # Separación del dipolo/tripolo/cuadripolo (m)

A = 1      # Amplitud (factor de proporcionalidad)

R = 10     # Distancia al punto de observación (m)
